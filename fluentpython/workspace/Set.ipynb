## 集合操作


```python
s = {i for i in range(10)}
z = {0, 1, 2}
c = {2, 3}
r = s & z & c
print('s & z & c : ', r)
print('s.intersection(z, c) == r : ', s.intersection(z, c) == r)
print('s : ', s)
s.intersection_update(z, c) # equivalent to s &= (z & c)
print('After s.intersection_update(z, c), s : ', s)
```

- 集合中的元素必须是可散列的
- 集合很消耗内存
- 可以很高效地判断元素是否存在于某个集合
- 元素的次序取决于被添加到集合里的次序
- 往集合里添加元素，可能会改变集合里已有元素的次序
